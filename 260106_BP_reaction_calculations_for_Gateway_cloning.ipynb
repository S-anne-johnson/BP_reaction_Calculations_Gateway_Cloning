```python

#Input: spreadsheet that contains gene names (titled "Gene name"), DNA sequences (titled "BP cloning sequence"), and measured concentrations of the samples (titled "Concentration from Nanodrop (ng/uL)")

#The script will perform the following process:

  #Add column for the ng needed to get 25 fmol gene fragment.
      #ng needed to get 25 fmol gene fragment = (length of DNA sequence) x 0.0165

  #Add a dilution column. Add a column for the uL needed to pipet of the diluted or undiluted DNA sample.
      #Divide required nanograms by the concentration.
        #If the volume is approximately between 1-2 uL, this is the volume value. The dilution value is 'None'
        #Otherwise, if the volume is greater than 2 uL: make this the volume value anyway, but print an error message that the sample is too dilute. The dilution value is 'None'
        #Otherwise, if the volume is less than 1 uL: make a dilution (try 1/2X dilution first, then 1/3X, etc).
          #Add the dilution value to the dilution column (ex. 1/2X). Add the volume value to the volume column.

  #Outputs a spreadsheet containing the required dilutions and sample volumes to add for the BP cloning reaction.
```

```python
import pandas as pd

sheet = "Example_input_sheet"
gene_fragments = pd.read_excel(sheet + ".xlsx")


gene_fragments
gene_fragments["Ng required for 25 fmol"] = ""
gene_fragments["Dilution for BP reaction"] = ""
gene_fragments["Volume for BP reaction"] = ""
```

```python
for index in gene_fragments["BP cloning sequence"].index:
  seq = gene_fragments.loc[index, "BP cloning sequence"]
  conc = gene_fragments.loc[index, "Concentration from Nanodrop (ng/uL)"]

  required_ng = len(seq)*0.0165
  gene_fragments.loc[index, "Ng required for 25 fmol"] = required_ng
  vol = required_ng / conc
  if vol >= 0.9 and vol <= 2.1:
    gene_fragments.loc[index,"Dilution for BP reaction"] = "None"
    gene_fragments.loc[index,"Volume for BP reaction"] = round(vol,1)
  elif vol > 2.1:
    gene_fragments.loc[index,"Dilution for BP reaction"] = "None"
    gene_fragments.loc[index,"Volume for BP reaction"] = round(vol,1)
    print(f"ERROR: sample {gene_fragments.loc[index,"Gene name"]} is too dilute (required volume is {vol} uL). Consider concentrating the sample so that less than 2 uL can be added.")
  elif vol < 0.9:
    integer = [2,3,4,5,6,7,8,9,10]
    for i in integer:
      diluted_conc = conc/i
      diluted_vol = required_ng/diluted_conc
      if diluted_vol >= 0.9 and diluted_vol <= 2.1:
        gene_fragments.loc[index,"Dilution for BP reaction"] = f"1/{str(i)}X"
        gene_fragments.loc[index,"Volume for BP reaction"] = round(diluted_vol,1)
        break
```

```python
gene_fragments.to_excel(sheet + "_BP_calculations.xlsx",index = False)
```
